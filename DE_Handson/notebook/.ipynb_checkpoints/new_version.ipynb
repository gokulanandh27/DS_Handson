```python
import pandas as pd

df = pd.read_csv("C:\\Users\\GokulAnandhP\\MyJupyterProject\\DE_Handson\\data\\15_Investment Analytics.csv")
df.info()
df.head()
```

```python
df['entrepreneurs'] = df['entrepreneurs'].fillna('Unknown')
df['deal'] = df['deal'].astype(int)
df['askedfor'] = pd.to_numeric(df['askedfor'], errors='coerce')
df['valuation'] = pd.to_numeric(df['valuation'], errors='coerce')
```

```python
df['num_entrepreneurs'] = df['entrepreneurs'].apply(
    lambda x: len(str(x).split(','))
)
```

```python
df['ask_pct'] = (df['askedfor'] / df['valuation']) * 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(x='category', y='deal', data=df)
plt.xticks(rotation=90)
plt.title("Deal Success Rate by Category")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df,
    x='askedfor',
    y='valuation',
    hue='deal',
    alpha=0.7
)
plt.title("Asked Amount vs Valuation (Colored by Deal Outcome)")
plt.show()
```

```python
from scipy.stats import chi2_contingency

cont = pd.crosstab(df['category'], df['deal'])
chi2, p, dof, expected = chi2_contingency(cont)
p
```

```python
from scipy.stats import mannwhitneyu

success = df[df['deal'] == 1]['valuation']
fail = df[df['deal'] == 0]['valuation']

mannwhitneyu(success, fail, alternative='two-sided')
```

```python
from scipy.stats import ttest_ind

ttest_ind(
    df[df['deal'] == 1]['num_entrepreneurs'],
    df[df['deal'] == 0]['num_entrepreneurs']
)
```
