```python
import pandas as pd

df = pd.read_csv("../data/15_Investment Analytics.csv")
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df['deal'] = df['deal'].astype(int)   # convert true/false to 1/0
df['valuation'] = pd.to_numeric(df['valuation'], errors='coerce')
df['askedfor'] = pd.to_numeric(df['askedfor'], errors='coerce')
```

```python
df['num_entrepreneurs'] = df['entrepreneurs'].apply(lambda x: len(str(x).split(',')))
```

```python
df['sharks_involved'] = df[['shark1','shark2','shark3','shark4','shark5']].notna().sum(axis=1)
```

```python
df['valuation_to_ask_ratio'] = df['valuation'] / df['askedfor']
```

```python
!pip install seaborn matplotlib
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(x='category', y='deal', data=df)
plt.xticks(rotation=90)
plt.title("Deal Success Rate by Category")
```

```python
!pip install scipy
```

```python
from scipy.stats import chi2_contingency

cont = pd.crosstab(df['category'], df['deal'])
chi2, p, dof, expected = chi2_contingency(cont)

p
```

```python
from scipy.stats import mannwhitneyu

success = df[df['deal']==1]['valuation']
fail = df[df['deal']==0]['valuation']

mannwhitneyu(success, fail, alternative='two-sided')
```

```python
from scipy.stats import ttest_ind

ttest_ind(df[df['deal']==1]['num_entrepreneurs'],
          df[df['deal']==0]['num_entrepreneurs'])
```
