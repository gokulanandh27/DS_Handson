```python
import numpy as np
 
def train_test_split_manual(X, y, test_size, shuffle=True, random_state=None):
 
    X = np.array(X)
    y = np.array(y)
   
       
    n_samples = len(X)
    n_test = int(n_samples * test_size)
   
    indices = np.arange(n_samples)
   
    if shuffle:
        if random_state is not None:
            np.random.seed(random_state)
        np.random.shuffle(indices)
   
    test_indices = indices[:n_test]
    train_indices = indices[n_test:]
   
    return X[train_indices], X[test_indices], y[train_indices], y[test_indices]
 

 
 
```

```python
def linear_regression(x, y):
    X = np.array(x).flatten()
    Y = np.array(y).flatten() 

    y_mean = np.mean(Y)
    x_mean = np.mean(X) 
    beta_numerator=0
    beta_denominator=0
    for i in range(len(X)):
        beta_numerator += (X[i] - x_mean) * (Y[i] - y_mean)
        beta_denominator += (X[i] - x_mean) ** 2
    beta = beta_numerator / beta_denominator
    alpha = y_mean - beta * x_mean
    
    return alpha, beta
```

```python
 
X = [[1], [2], [3], [4], [5]]
y = [10, 20, 30, 40, 50]
 
X_train, X_test, y_train, y_test = train_test_split_manual(X, y, test_size=0.4, random_state=42)
 
print("X_train:", X_train)
print("X_test:", X_test)
print("y_train:", y_train)
print("y_test:", y_test)
```

```python
alpha, beta = map(float, linear_regression(X_train, y_train))
```

```python
y_pred = alpha + beta * X_test.flatten()
print("Predicted values for X_test:", y_pred)
```

```python

X_multi = np.array([
    [4, 1, 3],
    [9, 2, 1],
    [8, 3, 2],
    [8, 1, 1],
    [7, 1, 4],
    [5, 0, 2],
])
y_multi = np.array([52, 51, 50, 52, 70, 60])


def fit_multivariable_linear_regression(X, y):
    X = np.array(X, dtype=float)
    y = np.array(y, dtype=float)
    n_samples, n_features = X.shape
    
    y_mean = np.mean(y)
    x_means = np.mean(X, axis=0)
    
    betas = np.zeros(n_features)
    
    for j in range(n_features):
        beta_numerator = 0
        beta_denominator = 0
        for i in range(n_samples):
            beta_numerator += (X[i, j] - x_means[j]) * (y[i] - y_mean)
            beta_denominator += (X[i, j] - x_means[j]) ** 2
        betas[j] = beta_numerator / beta_denominator
    
    alpha = y_mean - np.sum(betas * x_means)
    
    return alpha, betas


alpha_multi, beta_multi = fit_multivariable_linear_regression(X_multi, y_multi)
print("Intercept (B0):", alpha_multi)
print("Coefficients (B1, B2, B3):", beta_multi)

X_new = np.array([
    [6, 1, 1],
    [6, 3, 2],
    [5, 2, 3],
    [9, 2, 3],
])
y_pred_new = alpha_multi + X_new @ beta_multi
print("Predicted values for new data:", y_pred_new)
```
